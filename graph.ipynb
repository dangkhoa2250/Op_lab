```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scienceplots
```

```python
df = pd.read_csv("kairo.txt", sep = "\t")
```

```python
df.head()
```

```python
time = df["time"]*1000
Vc1 = df["V(n002)"]
Vc2 = df["V(n005)"]
Vb1 = df["V(n004)"]
Vb2 = df["V(n003)"]
```

```python
plt.style.use(["science", "no-latex", "ieee"])
plt.rcParams["font.family"] = "Yu Mincho"
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(time, Vc1, color = "r", label = "コレクタ電圧VC1")
ax1.plot(time, Vb1, "--", color = "r", label = "ベース電圧VB1")
ax1.set_xlabel("時間[ms]", fontname = "Yu mincho")
ax1.set_ylabel("$V_{C1},V_{B1}$[V]", fontname = "Yu mincho")
ax2.set_ylabel("$V_{C2},V_{B2}$[V]", fontname = "Yu mincho")
ax2.plot(time, Vc2, color = "b", label = "コレクタ電圧VC2")
ax2.plot(time, Vb2,"--", color = "b",label = "ベース電圧VB2" )
ax1.set_xlim(0, 3.5)
ax1.set_ylim(-8,16)
ax2.set_ylim(-15,9)
ax1.legend(loc =3, fontsize = 6)
ax2.legend(loc =1, fontsize = 6)
plt.show()
```

```python
fig.savefig('kairo.eps',format='eps')
```
